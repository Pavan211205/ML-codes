```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('/content/Cardiotocographic.csv')
df
```

```python
# Data Cleaning and Preparation
# Check for missing values
df.isnull().sum()
```

```python
# Handle missing values (if any)
# For simplicity, we'll drop rows with missing values
df = df.dropna()
df
```

```python
#After dropping null values
df.shape
```

```python
# Check for inconsistencies in data types
df.dtypes
```

```python
# Detecting and treat outliers (using IQR method)
def detect_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    Lower_Limit = Q1 - (1.5 * IQR)
    Upper_Limit = Q3 + (1.5 * IQR)
    return column[(column < Lower_Limit ) | (column > Upper_Limit)]

outliers = {}
for col in df.select_dtypes(include=[np.number]).columns:
    outliers[col] = detect_outliers(df[col])

print("Outliers detected in each column are:")
for col, out in outliers.items():
    print(f"{col}: {len(out)} outliers")
```

```python
# Statistical Summary of the data
df.describe()
```

```python
# Data Visualization
# Histograms for numerical variables
df.hist(bins=30, figsize=(15, 10))
plt.suptitle("Histograms for Numerical Variables")
plt.show()
```

```python
# Boxplots for numerical variables
plt.figure(figsize=(15, 10))
sns.boxplot(data=df)
plt.title("Boxplots for Numerical Variables")
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(12, 8))
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Scatter plots for selected pairs of variables
sns.pairplot(df[['LB', 'AC', 'FM', 'UC', 'ASTV', 'MSTV']])
plt.suptitle("Pair Plots of Selected Variables", y=1.05)
plt.show()
```
