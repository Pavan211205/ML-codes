```python
import pandas as pd
df = pd.DataFrame([[50,80,60,30,20], [70,100,90,50,50]],index= ['Smart Thermostat', 'Smart Light'], columns=['Very Satified', 'Satisfied', 'Neutral', 'Unsatisfied','Very Unsatified'])
df
```

```python
sum_1=sum(df.loc['Smart Thermostat'])
print('Sum of Smart Thermostat',sum_1)
sum_2=sum(df.loc['Smart Light'])
print('Sum of Smart Light',sum_2)
total_sum=sum_1+sum_2
print('Total Sum',total_sum)
```

```python
from scipy import stats
Null_Hypothesis= 'There is no significant association between the type of device purchased and the customers satisfaction level.'
Alternate_Hypothesis = 'There is a statistically significant association between type of device purchased and the customers satisfaction level .'
print(stats.chi2_contingency(df))
chi2_statistic=stats.chi2_contingency(df)[0]
print('The chi-square statistic value is',chi2_statistic)
#Inverse Survival Function
chi2_critical = stats.chi2.isf(0.05,4)
print('The chi-square critical value is',chi2_critical)
p_value =stats.chi2_contingency(df)[1]
print('The p-value is',p_value)
if (chi2_statistic > chi2_critical):
  print(Alternate_Hypothesis)
  print('Also the p-value of',p_value, 'is lesser than significance level of 0.05, hence Alternate Hypothesis is true')
else:
  print(Null_Hypothesis)
  print('Also the p-value of',p_value, 'is greater than significance level of 0.05, hence Null Hypothesis is true')

#Coclusion:
#Since the Chi-Square statistic (5.6382) is less than the critical value (9.4877) and the p-value (0.2278) is greater than 0.05, we do not have sufficient evidence to conclude that customer satisfaction is significantly associated with the type of smart home device purchased. â€‹
```
