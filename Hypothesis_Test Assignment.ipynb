```python
sample_mean = 3050
theoretical_mean = 4000 #μ = 1000 + 5 * 600 = 4000
std_dev_sample_mean = 25 #σ/√n = (5 * 25) / √25 = 25
sample_size = 25
alpha = 0.05

# Calculate test statistic
t_statistic = (sample_mean - theoretical_mean) / std_dev_sample_mean
print('The t-statistical value is', t_statistic)

# Calculate p-value
p_value = 1 - stats.norm.cdf(t_statistic)
print('The p-value is',p_value)

#Calculate the critical value
critical_value = stats.norm.ppf(1 - alpha)
print('The critical value is',critical_value)

#Since the p value > the significance value 0.05 hence we fail to reject Null Hypothesis.

# Comparing t-statistical value with critical value.
if t_statistic > critical_value:
    print("Reject the null hypothesis.")
    print("There is sufficient evidence to support the claim that operating costs are higher.")
else:
    print("Fail to reject the null hypothesis.")
    print("There is not enough evidence to support the claim that operating costs are higher.")

#Conlusion:
#There is not enough evidence to support the restaurant owners' claim that the weekly operating costs are significantly higher than those predicted by the model.
```
