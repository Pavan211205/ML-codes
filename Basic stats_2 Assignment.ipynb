```python
import pandas as pd
heads=pd.read_excel('/content/Basic stats_2 solved Assignment.xlsx',sheet_name='HEADS')
print(heads)
import numpy as np
from scipy import stats
```

```python
#When Population standard deviation is given
#sample mean=1.238667
#sample std deviation=0.193164
#population std deviation=0.2
#sample size(n)=15
#Degrees of freedom=14
sample_mean=1.238667
print('The average value of sample is',sample_mean)
std_error=0.2/(np.sqrt(15))
print('The standard error is',std_error)
#To get the t-value for given required confidnce interval & specified degrees of freedom
stats.t.interval(0.99,df=14)

print('Lower Limit of 99% of CI',round(sample_mean+(stats.t.interval(0.99,df=14)[0])*(std_error),3))
print('Upper Limit of 99% of CI',round(sample_mean+(stats.t.interval(0.99,df=14)[1])*(std_error),3))

```

```python
avg_bal_int=stats.t.interval(0.99,14,loc=sample_mean,scale=std_error)
print('Average is expected between',np.round(avg_bal_int[0],3),'&',np.round(avg_bal_int[1],3))
```

```python
#When Population standard deviation is not given
#sample mean=1.238667
#sample std deviation=0.193164
sample_mean=1.238667
print('The average value of sample is',sample_mean)
sample_std_dev=heads['Print heads'].std()
print('The average value of sample is',sample_std_dev)
std_error=sample_std_dev/(np.sqrt(15))
print('The standard error is',std_error)
#To get the t-value for given required confidnce interval & specified degrees of freedom
stats.t.interval(0.99,df=14)

print('Lower Limit of 99% of CI',round(sample_mean+(stats.t.interval(0.99,df=14)[0])*(std_error),3))
print('Upper Limit of 99% of CI',round(sample_mean+(stats.t.interval(0.99,df=14)[1])*(std_error),3))
```

```python
avg_bal_int=stats.t.interval(0.99,14,loc=sample_mean,scale=std_error)
print('Average is expected between',np.round(avg_bal_int[0],3),'&',np.round(avg_bal_int[1],3))
```
